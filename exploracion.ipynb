# Exploración del dataset
    Aquí se analizará el dataset para posteriormente aplicar transfromaciones de ser necesario

```python
import pandas as pd
import numpy as np
import logging
from config import API_BASE_URL, API_TOKEN, EMAIL
from ingest import fetch_data, parse_data

PARAMS = {
    "email": EMAIL,
    "key": API_TOKEN,
    "type": "ecommerce",
    "rows": 1000,
    "format": "json"
}

CATEGORIES = 'categories'
BRANDS = 'brands'
SUPPLIERS = 'suppliers'
WAREHOUSES = 'warehouses'
PRODUCTS = 'products'
INVENTORY = 'inventory'
CUSTOMERS = 'customers'
PROMOTIONS = 'promotions'
ORDERS = 'orders'
ORDER_ITEMS = 'order_items'
REVIEWS = 'reviews'

logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

try:
    resp = fetch_data(API_BASE_URL, PARAMS, logger)
    tables = parse_data(resp)

    print(tables.keys())
except Exception as e:
    logger.error(f"An error occurred: {e}")
```
